### User input.

Give the user a way to give input to the Rec System, with questions!

```python
#Getting the user's input for time management
def get_timemanagement():
    timemanagement = input("Do you like journaling, lists or calendars (devide by comma if multiple, or add a new one)? [Type 'skip' to skip this question] ")
    timemanagement = " ".join(["".join(n.split()) for n in timemanagement.lower().split(',')])
    return timemanagement

def get_difficulty():
    difficulty = input("Do you like daily schedules? [Type 'daily schedule if yes', type skip' to skip this question] ")
    difficulty = " ".join(["".join(n.split()) for n in difficulty.lower().split(',')])
    return difficulty

def get_feedback():
    feedback = input("Is a new Time management method for school, work or personal? [devide with comma if multiple, 'skip' to skip this question] ")
    feedback = " ".join(["".join(n.split()) for n in feedback.lower().split(',')])
    return feedback

def get_focus():
    focus = input("Do you take a lot of breaks? [Type 'multiple breaks if yes' 'skip' to skip this question] ")
    focus = " ".join(["".join(n.split()) for n in focus.lower().split(',')])
    return focus

def get_keywords():
    keywords = input("Are there any keywords that we need to include in your search result for this interaction? (please devide with a comma), [Type 'skip' to skip this question] ")
    keywords = " ".join(["".join(n.split()) for n in keywords.lower().split(',')])
    return keywords

def get_searchTerms():
    searchTerms = [] 
    timemanagement = get_timemanagement()
    if timemanagement != 'skip':
        searchTerms.append(timemanagement)

    difficulty = get_difficulty()
    if difficulty != 'skip':
        searchTerms.append(difficulty)

    feedback = get_feedback()
    if feedback != 'skip':
        searchTerms.append(feedback)

    focus = get_focus()
    if focus != 'skip':
        searchTerms.append(focus)
        
    keywords = get_keywords()
    if keywords != 'skip':
        searchTerms.append(keywords)
  
    return searchTerms
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

def make_recommendation(df1=df1):
    new_row = df1.iloc[-1,:].copy() #creating a copy of the last row of the  #dataset, which we will use to input the user's input
    
  #grabbing the new wordsoup from the user
    searchTerms = get_searchTerms()  
    new_row.iloc[-1] = " ".join(searchTerms) #adding the input to our new row
  
  #adding the new row to the dataset
    df1 = df1.append(new_row)
  
  #Vectorizing the entire matrix!
    count = CountVectorizer(stop_words='english')
    df1['soup'] = df1['soup'].values.astype('U')
    count_matrix = count.fit_transform(df1['soup'])

  #running pairwise cosine similarity 
    cosine_sim2 = cosine_similarity(count_matrix, count_matrix) #getting a similarity matrix
  
  #sorting cosine similarities by highest to lowest
    sim_scores = list(enumerate(cosine_sim2[-1,:]))
    #sim_scores = sim_scores[0:9]
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  #matching the similarities to the movie titles and ids
    ranked_titles = []
    for i in range(1, 11):
        indx = sim_scores[i][0]
        ranked_titles.append([df1['title_id'].iloc[indx],df1['Title'].iloc[indx]]) #,df1['Description'].iloc[indx]
    
    return ranked_titles
    
    #match_indices = [i[0] for i in sim_scores]
    
    #return df1['Title'].iloc[match_indices]
```
